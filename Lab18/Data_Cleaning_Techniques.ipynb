# Data_Cleaning_Techniques

```python
df = pd.read_csv('./raw_data.csv', header=None)
df.columns = ['user_name', 'age', 'town']
df
```

```python
# replace all town NaN with Sofia:
# df.town.fillna(value='SOFIA')
df.fillna(value={'town':'SOFIA', 'age': df.age.median()})
```

```python
df
```

```python
df.bfill()
```

```python
df
```

```python
df.loc[3,:] = np.NaN
df
```

```python
df.dropna(axis=1, how='any')
```

```python
df
```

```python
df.loc[4,'town'] = 'Burgas'
df
```

```python
df.drop_duplicates(subset='user_name', keep='first')
```

```python
df
```

```python
df.replace(to_replace=['Ivan', 'Maria'], value=['11111', '2222'] )
```
