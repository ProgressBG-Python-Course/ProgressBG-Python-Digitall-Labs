```python
import pandas as pd
import numpy as np
```

## Basics

```python
prices_ds = pd.Series(
    [1.5, 2, 2.5, 3],
    index=["apples", "oranges", "bananas", "strawberries"]
)

suppliers_ds = pd.Series(
    ["supplier1", "supplier2", "supplier4", "supplier3"],
    index=["apples", "oranges", "bananas", "strawberries"]
)

fruits_df = pd.DataFrame({
    "prices": prices_ds,
    "suppliers": suppliers_ds
})

fruits_df
```

## Create DF from json

```python
json_file = 'https://raw.githubusercontent.com/geekcourses/JupyterNotebooksExamples/master/datasets/python_books/books.json'
books = pd.read_json(json_file)
books.head(3)
```

```python
csv_file = './datasets/books.csv'
books.to_csv(csv_file, index=False)
```
